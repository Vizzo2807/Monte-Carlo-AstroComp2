```python
import matplotlib.pyplot as plt
import numpy as np
```

 Crear un sample de 100 estrellas con masas entre 0.08 y 100 masas solares

```python
# Definir la función IMF de Kroupa
def imf_kroupa(m):
    if m <= 0.08:
        return m**(-0.3)
    elif 0.08 < m <= 0.5:
        return m**(-1.3)
    else:
        return m**(-2.3)
```

```python
# Definir el número de estrellas para cada muestra
num_estrellas_1 = 100
num_estrellas_2 = 1000
num_estrellas_3 = 10000
num_estrellas_4 = 100000
num_estrellas_5 = 1000000
```

```python
def generar_masas_por_imf(cantidad_estrellas, masa_min=0.08, masa_max=100):
    lista_masas = []
    while len(lista_masas) < cantidad_estrellas:
        # Genera una masa aleatoria dentro del rango especificado
        masa_actual = np.random.uniform(masa_min, masa_max)
        # Calcula la probabilidad basada en la IMF de Kroupa
        probabilidad = imf_kroupa(masa_actual)
        # Acepta la masa si el valor aleatorio está dentro de la probabilidad
        if np.random.random() < probabilidad:
            lista_masas.append(masa_actual)
    return np.array(lista_masas)


# Generar masas para cada muestra
masas_1 = generar_masas_por_imf(num_estrellas_1)
masas_2 = generar_masas_por_imf(num_estrellas_2)
masas_3 = generar_masas_por_imf(num_estrellas_3)
masas_4 = generar_masas_por_imf(num_estrellas_4)
masas_5 = generar_masas_por_imf(num_estrellas_5)

# Graficar la distribución de masas para la muestra más grande
plt.figure(figsize=(10, 6))
plt.hist(masas_5, bins=100, density=True, alpha=0.7, color='b')
plt.xlabel('Masa (M_sol)')
plt.ylabel('Densidad de Probabilidad')
plt.title('Distribución de Masas de Estrellas (IMF de Kroupa)')
plt.grid(True)
plt.show()
```

```python
# Crear los subplots para mostrar los histogramas uno al lado del otro
fig, axs = plt.subplots(1, 5, figsize=(20, 4), sharey=True)  # 1 fila, 5 columnas

# Generar los histogramas en cada subplot
axs[0].hist(masas_1, bins=50, color='b', alpha=0.7, density=True)
axs[0].set_title('100 stars')
axs[0].set_xlabel('Mass [$M_{\odot}$]')
axs[0].set_ylabel('Frecuencia')

axs[1].hist(masas_2, bins=50, color='g', alpha=0.7, density=True)
axs[1].set_title('1000 stars')
axs[1].set_xlabel('Mass [$M_{\odot}$]')

axs[2].hist(masas_3, bins=50, color='r', alpha=0.7, density=True)
axs[2].set_title('10000 stars')
axs[2].set_xlabel('Mass [$M_{\odot}$]')

axs[3].hist(masas_4, bins=50, color='c', alpha=0.7, density=True)
axs[3].set_title('100000 stars')
axs[3].set_xlabel('Mass [$M_{\odot}$]')

axs[4].hist(masas_5, bins=50, color='m', alpha=0.7, density=True)
axs[4].set_title('1000000 stars')
axs[4].set_xlabel('Mass [$M_{\odot}$]')

# Ajustar los espacios entre los subplots
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
# Constantes
t_sun = 10**10  # Lifetime of the Sun in years (10 billion years)


# Calcular el tiempo en la secuencia principal para cada masa
TMS_1 = t_sun / (masas_1**2.5)  # TMS en años
TMS_2 = t_sun / (masas_2**2.5)  # TMS en años
TMS_3 = t_sun / (masas_3**2.5)  # TMS en años
TMS_4 = t_sun / (masas_4**2.5)  # TMS en años
TMS_5 = t_sun / (masas_5**2.5)  # TMS en años

# Calcular la edad actual de las estrellas (10 Gyr menos el tiempo de nacimiento)
ages_1 = 10**9 * (10 - birth_times1)  # Edad en Gyr
ages_2 = 10**9 * (10 - birth_times2)  # Edad en Gyr
ages_3 = 10**9 * (10 - birth_times3)  # Edad en Gyr
ages_4 = 10**9 * (10 - birth_times4)  # Edad en Gyr
ages_5 = 10**9 * (10 - birth_times5)  # Edad en Gyr

# Determinar si la estrella es un remanente estelar (edad > TMS)
stellar_remnants1 = ages_1 > TMS_1
stellar_remnants2 = ages_2 > TMS_2
stellar_remnants3 = ages_3 > TMS_3
stellar_remnants4 = ages_4 > TMS_4
stellar_remnants5 = ages_5 > TMS_5

# Contar cuántas estrellas son remanentes y cuántas están vivas
num_remnants_1 = np.sum(stellar_remnants1)
num_alive_1 = num_estrellas_1 - num_remnants_1

num_remnants_2 = np.sum(stellar_remnants2)
num_alive_2 = num_estrellas_2 - num_remnants_2

num_remnants_3 = np.sum(stellar_remnants3)
num_alive_3 = num_estrellas_3 - num_remnants_3

num_remnants_4 = np.sum(stellar_remnants4)
num_alive_4 = num_estrellas_4 - num_remnants_4

num_remnants_5 = np.sum(stellar_remnants5)
num_alive_5 = num_estrellas_5 - num_remnants_5
```

```python
print(f"Estrellas vivas (en la secuencia principal): {num_alive_1}, {num_alive_2}, {num_alive_3}, {num_alive_4}, {num_alive_5}")
print(f"Remanentes estelares: {num_remnants_1}, {num_remnants_2}, {num_remnants_3}, {num_remnants_4}, {num_remnants_5}")
```

```python
# Clasificación de los remanentes estelares
def classify_remnant(mass):
    if mass > 20:
        return 'Black Hole'
    elif 8 <= mass <= 20:
        return 'Neutron Star'
    else:
        return 'White Dwarf'
```

```python
# Aplicar la clasificación a cada conjunto de remanentes
remnants_type_1 = np.array([classify_remnant(m) for m in masas_1[stellar_remnants1]])
remnants_type_2 = np.array([classify_remnant(m) for m in masas_2[stellar_remnants2]])
remnants_type_3 = np.array([classify_remnant(m) for m in masas_3[stellar_remnants3]])
remnants_type_4 = np.array([classify_remnant(m) for m in masas_4[stellar_remnants4]])
remnants_type_5 = np.array([classify_remnant(m) for m in masas_5[stellar_remnants5]])
```

## Edades

```python
import numpy as np

# Definir constantes
t_sun = 10**10  # Lifetime of the Sun in years (10 billion years)

# Funciones para calcular la vida en la secuencia principal y clasificar remanentes
def calculate_tms(mass):
    return t_sun / (mass**2.5)

def classify_remnant(mass):
    if mass > 20:
        return 'Black Hole'
    elif 8 <= mass <= 20:
        return 'Neutron Star'
    else:
        return 'White Dwarf'

# Función para procesar cada conjunto de masas y tiempos de nacimiento
def process_stellar_data(masas, birth_times, num_estrellas):
    # Calcular tiempo en la secuencia principal y edad actual de las estrellas
    TMS = calculate_tms(masas)
    ages = 10**9 * (10 - birth_times)  # Edad en años

    # Determinar si la estrella es un remanente estelar
    stellar_remnants = ages > TMS

    # Contar cuántas estrellas son remanentes y cuántas están vivas
    num_remnants = np.sum(stellar_remnants)
    num_alive = num_estrellas - num_remnants

    # Clasificación de los remanentes estelares
    remnants_type = np.array([classify_remnant(m) for m in masas[stellar_remnants]])

    # Inicializar listas para almacenar las masas y edades de cada tipo de remanente
    white_dwarfs_masses = []
    white_dwarfs_ages = []
    neutron_stars_masses = []
    neutron_stars_ages = []
    black_holes_masses = []
    black_holes_ages = []
    main_sequence_masses = []
    main_sequence_ages = []

    # Extraer las masas y edades según el tipo de remanente o si está en secuencia principal
    for mass, age, remnant in zip(masas, ages, stellar_remnants):
        if remnant:
            remnant_type = classify_remnant(mass)
            if remnant_type == 'White Dwarf':
                white_dwarfs_masses.append(float(mass))
                white_dwarfs_ages.append(float(age))
            elif remnant_type == 'Neutron Star':
                neutron_stars_masses.append(float(mass))
                neutron_stars_ages.append(float(age))
            elif remnant_type == 'Black Hole':
                black_holes_masses.append(float(mass))
                black_holes_ages.append(float(age))
        else:
            main_sequence_masses.append(float(mass))
            main_sequence_ages.append(float(age))

    # Devolver resultados
    return {
        'main_sequence_masses': np.array(main_sequence_masses),
        'main_sequence_ages': np.array(main_sequence_ages),
        'white_dwarfs_masses': np.array(white_dwarfs_masses),
        'white_dwarfs_ages': np.array(white_dwarfs_ages),
        'neutron_stars_masses': np.array(neutron_stars_masses),
        'neutron_stars_ages': np.array(neutron_stars_ages),
        'black_holes_masses': np.array(black_holes_masses),
        'black_holes_ages': np.array(black_holes_ages)
    }


result_1 = process_stellar_data(masas_1, birth_times1, num_estrellas_1)
result_2 = process_stellar_data(masas_2, birth_times2, num_estrellas_2)
result_3 = process_stellar_data(masas_3, birth_times3, num_estrellas_3)
result_4 = process_stellar_data(masas_4, birth_times4, num_estrellas_4)
result_5 = process_stellar_data(masas_5, birth_times5, num_estrellas_5)

# Acceder a los resultados
print("Main Sequence Masses:", result_1['main_sequence_masses'])
print("Main Sequence Ages:", result_1['main_sequence_ages'])
print("White Dwarfs Masses:", result_1['white_dwarfs_masses'])
print("White Dwarfs Ages:", result_1['white_dwarfs_ages'])
print("Neutron Stars Masses:", result_1['neutron_stars_masses'])
print("Neutron Stars Ages:", result_1['neutron_stars_ages'])
print("Black Holes Masses:", result_1['black_holes_masses'])
print("Black Holes Ages:", result_1['black_holes_ages'])
```
